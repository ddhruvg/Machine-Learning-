```python
import pandas as pd
import matplotlib.pyplot as plt

# Data from the original table
data = {
    'Temperature(°C)': [27.08, 31.032, 34.9, 36.95, 38.93, 40.9, 42.88, 42.88, 44.85, 44.85, 44.85],
    'Resistance(Ω)': [109, 111, 113, 114, 115, 116, 117, 117, 118, 118, 18]
}

df = pd.DataFrame(data)

# Remove the last outlier point (the last row)
df_filtered = df.drop(df.index[-1])

# Create the line plot with markers
plt.figure(figsize=(10, 6))
plt.plot(df_filtered['Temperature(°C)'], df_filtered['Resistance(Ω)'], color='blue', marker='o', linestyle='-')

# Label the axes and title
plt.title('Resistance vs. Temperature (Outlier Removed)')
plt.xlabel('Temperature (°C)')
plt.ylabel(r'Resistance ($\Omega$)')

# Adjust y-axis limits to better show the trend
plt.ylim(105, 120) # Focusing on the relevant range after removing the outlier

# Add a grid for better readability
plt.grid(True, linestyle='--', alpha=0.6)

# Set tight layout
plt.tight_layout()

# Save the plot
plot_filename = 'resistance_vs_temperature_no_outlier_line_regenerated.png'
plt.savefig(plot_filename)

print(f"Plot saved to {plot_filename}")
```
